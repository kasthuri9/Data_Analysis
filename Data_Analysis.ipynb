```python
df.drop(['company','agent'], axis= 1, inplace= True)
df.dropna(inplace= True)
#list of column that is getting dropped
#then removing any rows that have missing values across any columns
#axis=1 indicate column and axis=0 indicate rows
#inplace=True: this makes changes directly to the OG Dataframe
```

```python
df['adr'].plot(kind= 'box')
# box plot 
```

```python
df= df[df['adr']<5000]
# will contain only the rows where the 'adr' value is less than 5000
```

```python
#Data analysis and visualisation
```

```python
cancelled_perc= df['is_canceled'].value_counts(normalize= True)
cancelled_perc
#the cancelled_perc variable will shows the percentage of canceled (1) and not canceled (0)
```

```python
print(cancelled_perc)
plt.figure(figsize= (5,4))
plt.title('Reservation status count')
plt.bar(['Not cancelled','Cancelled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
```

```python
plt.figure(figsize= (8,4))
ax1=sns.countplot(x='hotel',hue= 'is_canceled',data=df, palette='Blues')
legend_labels,_= ax1.get_legend_handles_labels()
ax1.legend(legend_labels, ['Not Cancelled', 'Cancelled'], bbox_to_anchor=(1, 1))
plt.title('Reservation status in different hotels',size= 15)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.show()
```

```python
resort_hotel= df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
#filters where the hotel column is equal to Resort Hotel
#calculates the value in 'is_canceled' column
```

```python
city_hotel= df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel= city_hotel.groupby('reservation_status_date')[['adr']].mean()
#groups the x.hotel by reservation date and then calculates the mean of adr
```

```python
plt.figure(figsize=(20,10))
plt.title('Average Daily Rate in Resort and City Hotel', fontsize=10)
plt.plot(resort_hotel.index,resort_hotel['adr'],label= 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'],label='City Hotel')
plt.legend(fontsize= 15)
plt.show()
```

```python
df['month']= df['reservation_status_date'].dt.month
plt.figure(figsize= (18,8))
ax1= sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels,_= ax1.get_legend_handles_labels()
ax1.legend(legend_labels, ['Not Cancelled', 'Cancelled'], bbox_to_anchor=(1, 1))
plt.title('Reservation status per month',size= 15)
plt.xlabel('Month',size=15)
plt.ylabel('Number of reservations', size=15)
plt.show()
```

```python
plt.figure(figsize= (15,8))
plt.title('Average Daily Rate per month',fontsize=25)
sum_adr_by_month= df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x='month',y='adr', data=sum_adr_by_month)
plt.show
           
```

```python
cancelled_data= df[df['is_canceled']== 1]
top_10_country= cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(15,8))
plt.title('Top 10 countries with reservation cancelled')
plt.pie(top_10_country, autopct= '%.2f', labels= top_10_country.index)
plt.show()
```

```python
cancelled_df_adr= cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)


not_cancelled_data= df[df['is_canceled']==0]
not_cancelled_df_adr= not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize= (20,6))
plt.title('Average Daily Rate', fontsize=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='Not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='Cancelled')
plt.legend()
plt.show()
```
